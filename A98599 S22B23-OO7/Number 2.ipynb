```python
import matplotlib.pyplot as plt
import numpy as np
from timeit import time

#insertion sort
def insertion_sort(arr):
  for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and key < arr[j]:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j + 1] = key

#merge
def merge_sort(arr):
  if len(arr) <= 1:
    return arr
  mid = len(arr) // 2
  left = merge_sort(arr[:mid])
  right = merge_sort(arr[mid:])
  return merge(left, right)

def merge(left, right):
  result = []
  i = j = 0
  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      result.append(left[i])
      i += 1
    else:
      result.append(right[j])
      j += 1
  result.extend(left[i:])
  result.extend(right[j:])
  return result


array_sizes = np.arange(100, 1000, 100)
execution_times_insertion_sort = []
execution_times_merge_sort = []
for array_size in array_sizes:
  random_array = np.random.randint(1, 100, array_size)

  # Measure execution time for each sorting algorithm
  start_time = time.time()
  insertion_sort(random_array.copy())
  insertion_time = time.time() - start_time

  start_time = time.time()
  merge_sort(random_array.copy())
  merge_time = time.time() - start_time

  # Store execution times
  execution_times_insertion_sort.append(insertion_time)
  execution_times_merge_sort.append(merge_time)

# Plotting the execution time
plt.plot(array_sizes, execution_times_insertion_sort, color='green', label='Insertion Sort')
plt.plot(array_sizes, execution_times_merge_sort, color='orange', label='Merge Sort')
plt.xlabel('Array Size (n)')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time of Sorting Algorithms')
plt.legend()
plt.show()

```
